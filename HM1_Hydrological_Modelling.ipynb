```r
library(airGRteaching)
library(sp)
```

```r
#Download catchment data
UpK1<-readRDS(gzcon(url('https://github.com/SiuSunChun/HydroConceptualModel/raw/main/UpK1.rds')))
UpK2<-readRDS(gzcon(url('https://github.com/SiuSunChun/HydroConceptualModel/raw/main/UpK2.rds')))
```

```r
##Plot catchment
plot(UpK1,border='black')
plot(UpK2,add=T,col='blue')
box()
axis(1)
axis(2)
title('39043 - Kennet at Knighton',lin=0.5)
grid(col = "lightgray", lty = 2, lwd =2)
```

```r
##Load  Kennet at Knighton
##https://nrfa.ceh.ac.uk/data/station/info/39043
MYD1m<-read.csv('https://raw.githubusercontent.com/SiuSunChun/HydroConceptualModel/refs/heads/main/D01_Kennet.csv')
```

```r
# Data processing for GR2M
prep <- PrepGR(DatesR     = as.POSIXlt(as.Date(MYD1m$Date)),
               Precip     = MYD1m$Ptot,
               PotEvap    = MYD1m$Evap, 
               Qobs       = MYD1m$Discharge,
               HydroModel = "GR2M", 
               CemaNeige  = FALSE)

plot(prep)
```

```r
###Calabration 
CAL <- CalGR(PrepGR = prep, CalCrit = "KGE2",
             WupPer = NULL, CalPer = c("1970-12-01", "1990-12-01"))

plot(CAL)
```

```r
###Simulation 
SIM <- SimGR(PrepGR = prep, Param = CAL, EffCrit = "KGE2",
             WupPer = NULL, SimPer = c("1970-01-01", "2020-12-01"))

plot(SIM)
```
